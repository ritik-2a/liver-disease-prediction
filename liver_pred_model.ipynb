Split the DataFrame df into features (X) and the target variable (y), typically in preparation for machine learning tasks.

X = df.drop('Dataset', axis=1):

* This drops the Dataset column from the DataFrame df, leaving all other columns as features.
* axis=1 tells pandas to drop the column (axis 0 is for rows, and axis 1 is for columns).
* The remaining columns are stored in the variable X, which represents the feature matrix (the independent variables used for prediction).

```python
y=df.Dataset
X=df.drop('Dataset', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

Split the dataset into training and testing sets for machine learning purposes. It divides the feature matrix X and the target variable y into separate training and testing sets, allowing you to evaluate the performance of your model on unseen data.



```python
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size=0.1, random_state=0, stratify=y)
```

calculate the mean and standard deviation of the features in the training dataset (X_train). These values are often used for feature scaling (standardization), which is a common preprocessing step in machine learning.



```python
train_mean = X_train.mean()
train_std = X_train.std()
```

standardization on the features in X_train, and then calls the describe() method to view summary statistics of the standardized data.

X_train = (X_train - train_mean) / train_std:

* This standardizes each feature in X_train by subtracting the mean and dividing by the standard deviation.
* Standardization transforms the data so that each feature has a mean of 0 and a standard deviation of 1.
* This is done for each feature (i.e., each column in X_train), where train_mean is the mean of the training data and train_std is the standard deviation.

```python
X_train = (X_train - train_mean) / train_std
X_train.describe()
```

    Standardize the test dataset (X_test) using the mean and standard deviation calculated from the training dataset (X_train).
    This ensures that the test data is scaled in the same way as the training data, which is a crucial step in machine learning to maintain consistency.

X_test = (X_test - train_mean) / train_std:

* Here, the mean (train_mean) and standard deviation (train_std) of the training dataset (X_train) are applied to the test dataset (X_test).
* This ensures that the test data is transformed using the same scaling parameters as the training data.
* It's important that the test set is standardized using the training set statistics, as this prevents the model from having knowledge of the test set during training.


```python
X_test = (X_test - train_mean) / train_std
X_test.describe()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

make predictions with the Logistic Regression model (lr) on the test dataset (X_test).

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

1. accuracy_score(y_train, lr.predict(X_train)):

* This line computes the accuracy of the logistic regression model on the training data (X_train) by comparing the true labels (y_train) with the predicted labels (lr.predict(X_train)).
* Accuracy is the proportion of correctly predicted labels out of all predictions.

2. lr_acc = accuracy_score(y_test, lr.predict(X_test)):
* This line computes the accuracy of the model on the test data (X_test), where y_test is the true target labels and lr.predict(X_test) gives the predicted labels for the test set.
* The result is stored in the variable lr_acc.

3. confusion_matrix(y_test, lr.predict(X_test)):
* The confusion matrix is computed, which shows the number of true positive, false positive, true negative, and false negative predictions.
* It is a great way to visually inspect the performance of a classification model, especially for imbalanced classes.

4. classification_report(y_test, lr.predict(X_test)):
* The classification report provides several important classification metrics,

```python
print(accuracy_score(y_train, lr.predict(X_train)))
lr_acc = accuracy_score(y_test, lr.predict(X_test))
print(lr_acc)
print(confusion_matrix(y_test, lr.predict(X_test)))
print(classification_report(y_test, lr.predict(X_test)))
```

computes the F1 score, which is the harmonic mean of precision and recall.

It is computed as:
F1= (2×Precision×Recall) / (Precision+Recall)


```python
from sklearn.metrics import f1_score
lr_f1 = f1_score(y_test, lr.predict(X_test))

print(f"F1 Score: {lr_f1:.2f}")
```

```python
from sklearn.metrics import precision_score
lr_precision = precision_score(y_test, lr.predict(X_test))

print(f"Precision: {lr_precision:.2f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)
```

    predictions using the K-Nearest Neighbors (KNN) model that was previously trained on the training data (X_train and y_train). Specifically, it predicts the class labels for the instances in the test data (X_test).
    The result will be an array of predicted labels corresponding to each test sample.




```python
knn.predict(X_test)
```

Evaluate the performance of the Logistic Regression (LR) model and the K-Nearest Neighbors (KNN) model. It computes accuracy, confusion matrix, and a detailed classification report for the KNN model on the test dataset.

```python
print("Accuracy Score: ",accuracy_score(y_train, lr.predict(X_train)))

knn_acc = accuracy_score(y_test, knn.predict(X_test))
print("Knn_Accuracy: ",knn_acc)


print(confusion_matrix(y_test, knn.predict(X_test)))

print(classification_report(y_test, knn.predict(X_test)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Confusion matrix for Logistic Regression
lr_cm = confusion_matrix(y_test, lr.predict(X_test))

# Confusion matrix for KNN
knn_cm = confusion_matrix(y_test, knn.predict(X_test))

# Plotting Logistic Regression Confusion Matrix
plt.figure(figsize=(12, 5))

# Logistic Regression confusion matrix heatmap
plt.subplot(1, 2, 1)
sns.heatmap(lr_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], yticklabels=['Class 1', 'Class 2'])
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

# KNN confusion matrix heatmap
plt.subplot(1, 2, 2)
sns.heatmap(knn_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], yticklabels=['Class 1', 'Class 2'])
plt.title('KNN Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

plt.tight_layout()
plt.show()
```
